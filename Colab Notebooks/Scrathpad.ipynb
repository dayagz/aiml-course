```python
data = {'customer': [101,102,103,104],
        'category': ['cat2','cat2','cat1','cat3'],
        'important': ['yes','no','yes','yes'],
        'sales': [123,52,214,663]}
dataframe = pd.DataFrame(data)
print(dataframe)

print(dataframe['category'].value_counts())

print(dataframe.sort_values(by='sales',ascending = False))

print(dataframe.loc[[1,2],['customer','sales']])
```

```python
data = {'customer': [101,102,103,104,101,103],
        'category': ['cat2','cat2','cat1','cat3','cat2','cat1'],
        'important': ['yes','no','yes','yes','yes','yes'],
        'sales': [123,52,214,663,204,453]}

dataframe = pd.DataFrame(data)
print(dataframe)
print(dataframe.groupby(['customer'])['sales'].mean())
```
